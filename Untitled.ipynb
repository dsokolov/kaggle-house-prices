```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
dataset_df = pd.read_csv('./datasets/train.csv')
```

```python
dataset_df.info()
```

```python
dataset_df.describe()
```

```python
msk = np.random.rand(len(dataset_df)) < 0.8
train_df = dataset_df[msk]
test_df = dataset_df[~msk]
```

```python
feature_columns = [
    tf.feature_column.numeric_column(key="LotArea", dtype=tf.int64)
]
```

```python
estimator = tf.estimator.DNNRegressor(
    hidden_units=[10,10],
    feature_columns=feature_columns,
    model_dir='./output/logs/'
)
```

```python
input_fn = tf.estimator.inputs.pandas_input_fn(
    x=train_df,
    y=train_df[['SalePrice']],
    shuffle=True
)

estimator.train(input_fn=input_fn, steps=100)
```
